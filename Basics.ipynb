```python
years = []

for row in data:
    years.append(row[1])
years
```

```python
year_counts = {}

for y in years:
    if y not in year_counts:
        year_counts[y] = 1
    else:
        year_counts[y] += 1
year_counts
```

```python
import datetime 

dates = []
for row in data:
    date = datetime.datetime(year= int(row[1]), month = int(row[2]), day = 1)
    dates.append(date)
dates[:5]
    
```

```python
date_counts = {}

for date in dates:
    if date not in date_counts:
        date_counts[date] = 1
    elif date in date_counts:
        date_counts[date] += 1

date_counts
```

```python
def column_count(column_number):
    colValues = []
    for row in data:
        colValues.append(row[column_number])

    colCounts = {}

    for value in colValues:
        if value not in colCounts:
            colCounts[value] = 1
        elif value in colCounts:
            colCounts[value] += 1
    return colCounts
```

```python
census = list(csv.reader(open('census.csv', 'r')))
census
census[0][-7]
```

```python
#dict that maps each key from race_counts to pop race count in census
mapping = {
    'Asian/Pacific Islander': int(census[1][-2]) + int(census[1][-3]),
    'Black': int(census[1][-5]),
    'Native American/Native Alaskan': int(census[1][-4]),
    'Hispanic': int(census[1][-6]),
    'White': int(census[1][-7])
}
mapping
```

```python
race_per_hundredk = {}

for race in race_counts:
    race_per_hundredk[race] = (race_counts[race] / mapping[race]) * 100000
race_per_hundredk
    
    
```

```python
intents = []
for row in data:
    intents.append(row[3])

races = []
for row in data:
    races.append(row[7])

homicide_race_per_hundredk = {}

for i, race in enumerate(races):
    if intents[i] == 'Homicide':
        if race not in homicide_race_per_hundredk:
            homicide_race_per_hundredk[race] = 1
        elif race in homicide_race_per_hundredk:
            homicide_race_per_hundredk[race] += 1
            
for race in homicide_race_per_hundredk:
    homicide_race_per_hundredk[race] = 100000 * (homicide_race_per_hundredk[race]/ mapping[race])

homicide_race_per_hundredk
```
